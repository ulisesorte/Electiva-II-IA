```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Definir los datos cualitativos sobre las prefencias del color
datos_cualitativos = {
    'Color' : ['Azul', 'Azul', 'Amarillo', 'Rojo', 'Rojo', 'Amarillo', 'Azul'],
    'Preferencia' : ['Alta', 'Alta', 'Baja', 'Media', 'Media', 'Baja', 'Alta']
}

# Crear el dataframe
df_cualitativos = pd.DataFrame(datos_cualitativos)

# Graficar con plt
plt.figure(figsize=(12,5))

# Creacion de subgrafica por filas y columnas
plt.subplot(1,2,1)
sns.countplot(x='Color', data=df_cualitativos, palette='viridis')
plt.title('Frecuencia de colores')
plt.xlabel('Colores')
plt.ylabel('Frecuencia')

# Creacion de subgrafica 2
plt.subplot(1,2,2)
sns.countplot(x='Preferencia', data=df_cualitativos, palette='magma')
plt.title('Frecuencia de preferencias')
plt.xlabel('Preferencia')
plt.ylabel('Frecuencia')
```

```python
datos_cuantitativos = { 
    'Edades' : [20, 50, 30, 20, 55, 60, 75],
    'Ingresos' : [3000, 1000, 5000, 2000, 1500, 3500, 1200]
}

# Creacion de df para cuantitativos
df_cuantitativos = pd.DataFrame(datos_cuantitativos)

plt.figure(figsize=(12, 5))

# Histograma de edades
plt.subplot(1, 2, 1)
sns.histplot(df_cuantitativos['Edades'], bins=5, kde=True, color='blue')
plt.title('Distribuci√≥n de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

```

# Medidas de Tendencia Central


```python
import statistics
# import matplotlib.pyplot as plt

edades = [20,50,36,55,25,85,41]

media = statistics.mean(edades)
mediana = statistics.median(edades)
moda = statistics.mode(edades)

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

# Graficar los datos
plt.hist(edades, bins=5, edgecolor='black', color='skyblue')
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana}')
plt.axvline(moda, color='purple', linestyle='dashed', linewidth=2, label=f'Moda: {moda}')
plt.title('Distribuci√≥n de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
